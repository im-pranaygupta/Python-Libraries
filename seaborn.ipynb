Seaborn : Data Visualization Library


```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Tips Dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
#Setting the Default theme
sns.set_theme()
#Visualizing the data
sns.relplot(data=tips, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')
```

```python
# IRIS Dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_width', hue='species')
```

```python
#TITANIC Dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.shape
```

```python
sns.countplot(x='class', data=titanic)
```

```python
sns.countplot(x='survived', data=titanic)
```

```python
#Bar Chart
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

```python
# Digits dataset
from sklearn.datasets import load_digits
digits_dataset = load_digits()
print(digits_dataset)
```

```python
digits = pd.DataFrame(digits_dataset.data, columns=digits_dataset.feature_names)
digits['Target'] = digits_dataset.target
digits.head()
```

```python
sns.displot(digits['Target'])
```

Correlation

```python
correlation = digits.corr()

# Heat Map
plt.figure(figsize=(20,20))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
