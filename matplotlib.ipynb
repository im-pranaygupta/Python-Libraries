```python
#MATPLOTLIB : Used for making plots
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Plotting the data
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
# Adding Title and Labels
plt.plot(x,y)
plt.xlabel('Angle (in radians)')
plt.ylabel('Sine Value')
plt.title('Sine Graph')
plt.show()
```

```python
# Parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
# Multiple Graphs
x = np.linspace(-5,5,50)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,'g')
plt.plot(x,z,'b')
plt.show()
```

```python
# Bar Plot
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100,20,50,70,10]
plt.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('No. of people')
plt.show()
```

```python
#Pie Chart
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100,20,50,60,10]
plt.pie(people,labels=languages, autopct='%1.1f%%')
plt.show()
```

```python
# Scatter Plot
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)

fig3 = plt.figure()
ax = fig3.add_axes([0,0,1,1])
ax.scatter(x,y,color ='g')
ax.scatter(x,z,color ='r')
plt.show()
```

```python
#3D Plot
fig4 = plt.figure()
ax = plt.axes(projection = '3d')
x = 20*np.random.random(100)
y = np.sin(x)
z = np.cos(x)

ax.scatter(x,y,z,c=x, cmap='Blues')
plt.show()
```
